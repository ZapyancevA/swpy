## Задание 1. Калькулятор скидок
1. Вы пишете модуль для кассового аппарата в магазине. Напишите функцию apply_discount, которая:

2. Принимает два аргумента, price (цена) и discount (процент скидки).

3. Установите значение скидки по умолчанию — 10%.

4. Функция должна возвращать итоговую стоимость товара.

**Проверка:**

1. Вызовите функцию, передав только цену (1000).

2. Вызовите функцию, передав цену (1000) и скидку (25).

```python
def apply_discount(price: float, discount: float = 10) -> float:

    final_price = price * (100 - discount) / 100
    return round(final_price, 2)

# Проверка работы функции
print("Кассовый аппарат - расчет стоимости со скидкой")
print("-" * 40)

# Вызов только с ценой (скидка по умолчанию 10%)
price1 = 1000
result1 = apply_discount(price1)
print(f"Товар 1: цена {price1} руб.")
print(f"Скидка по умолчанию (10%): {result1} руб.")
print()

# Вызов с ценой и скидкой 25%
price2 = 1000
discount2 = 25
result2 = apply_discount(price2, discount2)
print(f"Товар 2: цена {price2} руб., скидка {discount2}%")
print(f"Итоговая стоимость: {result2} руб.")
print("-" * 40)

# Дополнительные тесты
print("\nДополнительные примеры:")
print(f"Цена 500 руб., скидка 15%: {apply_discount(500, 15)} руб.")
print(f"Цена 2000 руб., скидка по умолчанию: {apply_discount(2000)} руб.")
print(f"Цена 750 руб., скидка 5%: {apply_discount(750, 5)} руб.")
```

## Задание 2. Чистый список (Работа со списками)
Данные из форм регистрации часто приходят с ошибками (лишние пробелы, разный регистр). Напишите функцию clean_names(names_list):

1. Функция принимает список строк (имен).

2. Она должна пройтись по списку, убрать пробелы по краям и привести имена к формату "Имя" (первая заглавная, остальные строчные).

3. Функция возвращает новый, исправленный список.

Пример входных данных: `[' алексей', 'МАРИЯ ', ' иВан ']`

```python
def clean_names(names_list: list) -> list:
    return [name.strip().lower().capitalize() for name in names_list]

# Проверка
input_names = [' алексей', 'МАРИЯ ', ' иВан ']
print(clean_names(input_names))  # ['Алексей', 'Мария', 'Иван']
```

## Задание 3. Счёт в ресторане (*args)
Официант вводит цены за каждое блюдо. Мы не знаем заранее, сколько блюд заказал гость. Напишите функцию calculate_bill(*items):

1. Функция принимает любое количество цен (чисел).

2. Используйте цикл for, чтобы сложить все переданные значения.

3. Верните итоговую сумму.

**Проверка:** Вызовите функцию для заказов: (250, 500, 100) и (1200, 300).

```python
def calculate_bill(*items: float) -> float:
    return round(sum(items), 2)

# Проверка
print(calculate_bill(250, 500, 100))    # 850
print(calculate_bill(1200, 300))        # 1500
```

## Задание 4. Логирование проблем (**kwargs)
Вы создаете систему тикетов для техподдержки. Напишите функцию `log_issue(user, details)`:

1. user — обязательный аргумент (имя пользователя).

2. details — произвольные детали (например, `device="Android"`, `priority="High"`).

3. Функция должна выводить на экран: "Пользователь` [имя]` сообщил о проблеме." и перечислять все переданные детали из kwargs.

**Пример вызова:** `log_issue("Артем", device="iPhone", error="404")`

```python
def log_issue(user, **details):
    print(f"Пользователь {user} сообщил о проблеме.")
    for key, value in details.items():
        print(f"  {key}: {value}")

# Проверка
log_issue("Артем", device="iPhone", error="404")
```

## Задание 5. Валидатор пароля (Type Hinting & Docstrings)
Нужно проверить, подходит ли пароль под требования безопасности. Напишите функцию is_strong_password:

1. Добавьте аннотацию типов: аргумент password (строка), возвращаемое значение (bool).

2. Добавьте Docstring (описание функции в тройных кавычках).

3. Функция должна возвращать True, если длина пароля 8 и более символов, иначе False.

```python
def is_strong_password(password: str) -> bool:
    """
    Проверяет, соответствует ли пароль требованиям безопасности.

    Требование: длина пароля должна быть не менее 8 символов.

    Параметры:
        password: строка с паролем для проверки

    Возвращает:
        bool: True если пароль надежный (длина >= 8), иначе False
    """
    return len(password) >= 8

# Проверка работы функции
print("Проверка надежности пароля")
print("-" * 30)

# Тестовые пароли
test_passwords = [
    "qwerty",        # 6 символов
    "1234567",       # 7 символов
    "12345678",      # 8 символов
    "StrongPass9",   # 10 символов
    "",              # пустой пароль
    "a" * 20         # 20 символов
]

for pwd in test_passwords:
    result = is_strong_password(pwd)
    status = "Надежный" if result else "Слабый"
    print(f"Пароль: '{pwd}' ({len(pwd)} симв.) -> {status}")
```
