## Часть 1. Цикл for и функция range()
Когда мы знаем, сколько раз нужно повторить действие, мы используем `for`. Часто он работает в паре с `range(start, stop, step)`.

**Синтаксис:**

```python
# range(5) создает последовательность: 0, 1, 2, 3, 4
for i in range(5):
    print(f"Повторение номер {i}")
```

***Обратите внимание: stop (5) не включается!***

## Часть 3. Цикл `while` (Пока...)
Когда мы не знаем, сколько раз нужно повторить действие (например, пока пользователь не введет правильный пароль), мы используем `while`.

**Синтаксис:**

```python
x = 0
while x < 5:    # Проверка условия ПЕРЕД каждым шагом
    print(x)
    x = x + 1   # Обязательно меняем x, иначе цикл будет вечным!
```

## Задание 3. «Копилка»
1. Напишите программу, которая помогает накопить на мечту.

2. Пусть цель накопления `(goal) = 1000` рублей.

3. Текущая сумма `(current) = 0.`

4. Запустите цикл `while`, который работает, пока `current < goal`.

5. Внутри цикла спрашивайте у пользователя: `"Сколько откладываем?"` и добавляйте это число к `current`.

6. Выводите текущий баланс на каждом шаге.

7. Когда цикл закончится (сумма набрана), выведите: `"Нужная сумма накоплена!"`.

```python
current = 0
goal = 1000
while current < goal:
  yfr = int (input("Сколько откладываем?"))
  current = current + yfr
  print(f"Текущий баланс: {current}")

print("Нужная сумма накоплена!")


```

## Часть 4. Итоговое задание
**Задача**: «Генератор статистики» Дан список оценок студента: `marks = [3, 4, 5, 2, 4, 3, 5, 5, 2]`

1. Используя только циклы (без функций `sum()` или `count()`):

2. Посчитайте количество пятерок.

3. Посчитайте сумму всех оценок (аккумулятор).

4. Вычислите средний балл (сумма / количество).

5. Выведите результаты на экран.

```python
marks = [3, 4, 5, 2, 4, 3, 5, 5, 2]

fives_count = 0
total_sum = 0
total_count = 0

for mark in marks:
    if mark == 5:
        fives_count += 1
    total_sum += mark
    total_count += 1

average = total_sum / total_count

print(f"Количество пятерок: {fives_count}")
print(f"Сумма всех оценок: {total_sum}")
print(f"Средний балл: {average}")
```
