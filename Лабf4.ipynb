## Часть 1. Анатомия функции
В программировании часто приходится делать одно и то же действие много раз. Чтобы не копировать код (Ctrl+C / Ctrl+V), мы используем функции.

**Синтаксис:**

```python
def my_function():      # 1. Объявление или по другому заголовок функции
    print("Привет!")    # 2. Тело функции (всегда необходимо добавлять отступы так же как и у циклов)
    print("Я работаю.")

# 3. Вызов функции (без этого код внутри не сработает)
my_function()
my_function()
```

**Важно**: Код внутри `def` не выполняется, пока вы не напишете имя функции со скобками `()` в основной части программы.

## Часть 2. Параметры (Аргументы)
Функция становится полезной, когда она может работать с разными данными. Данные передаются в скобках. Переменные в скобках называются параметрами (или аргументами).

**Пример:**

```python
def greet_user(name):          # name — это параметр
    print(f"Привет, {name}!")

greet_user("Алексей")          # "Алексей" попадает в name
greet_user("Мария")            # "Мария" попадает в name
```

Можно передавать несколько аргументов через запятую:

```python
def multiply_visual(a, b):
    print(f"{a} умножить на {b} равно {a * b}")

multiply_visual(5, 3)
```

## Часть 3. Самое важное: return или print?
Это главная ошибка новичков.

`print` — просто показывает картинку на экране. Программа не запоминает результат.

`return` — возвращает результат в программу, чтобы его можно было сохранить в переменную и использовать дальше.

После срабатывания `return` функция мгновенно прекращает работу (выходит).

**Сравните:**

```python
# ПЛОХО (Только печатает)
def add_print(a, b):
    print(a + b)

res = add_print(2, 2)
# В переменной res сейчас пусто (None), математика потеряна.

# ХОРОШО (Возвращает данные)
def add_return(a, b):
    return a + b

result = add_return(2, 2) # Результат (4) сохранился в result
final = result * 10       # Мы можем работать с числом дальше!
print(final)              # Выведет 40
```

1. Напишите функцию `get_vat(price)`, которая принимает цену товара.

2. Функция должна посчитать НДС (20% от цены) и вернуть (`return`) это число.

3. В основной программе запросите у пользователя цену (`input`).

4. Вызовите функцию, сохраните результат в переменную tax.

#№ Часть 4. Аргументы по умолчанию
Иногда мы хотим, чтобы у аргумента было значение "на всякий случай", если пользователь его не указал.

**Пример:**

```python
def connect_to_db(ip, port=3306): # port по умолчанию 3306
    print(f"Подключаюсь к {ip} через порт {port}")

connect_to_db("192.168.0.1")        # Используется стандартный порт 3306
connect_to_db("192.168.0.1", 8080)  # Стандартный заменяется на 8080
```

## Задание 4
Напишите функцию `email_generator(name, domain="gmail.com")`.

Она должна принимать имя пользователя и домен.

Если домен не указан, использовать `gmail.com`.

Функция должна возвращать готовый email (например, `user@gmail.com`).

Продемонстрируйте два вызова: один с указанием домена (например, `mail.ru`), второй — без.

```python
def email_generator(name, domain="gmail.com"):
    return name + "@" + domain

# Демонстрация
print(email_generator("user", "mail.ru"))     # user@mail.ru
print(email_generator("user"))                 # user@gmail.com
```
