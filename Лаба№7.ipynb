**Цель**: Освоить управление потоком (`break`, `continue`), вложенные циклы и создание бесконечных меню. Это основа для создания игр и сложных алгоритмов.

## Часть 1. Операторы break и continue
Иногда нужно нарушить ход цикла.

`break` — мгновенно прерывает цикл (выходит из него).

`continue` — пропускает текущий шаг и переходит к следующему.

**Пример:**

```python
for i in range(10):
    if i == 5:
        print("Встретили 5, пропускаем!")
        continue   # Прыгает обратно в начало цикла
    if i == 8:
        print("Встретили 8, хватит!")
        break      # Полностью убивает цикл
    print(i)
```

## Задание 1. «Цензор»
Есть список слов: `words = ["яблоко", "питон", "СТОП", "кодить", "солнце"]`.

Запустите цикл по этому списку.

Если слово совпадает со `"СТОП"` — нужно немедленно прекратить работу цикла (`break`).

Выводите все слова до слова `"СТОП"`.

```python
words = ["яблоко", "питон", "СТОП", "кодить", "солнце"]
for x in words:
  if x == 'СТОП':
    print("-ГОЙ? \n - да \n - не пропускаем")
    break
```

## Часть 2. Вечный цикл (`while True`)
Очень полезная конструкция для создания интерактивных меню. Цикл крутится вечно, пока внутри не сработает `break`.

**Схема меню:**

```python
while True:
    cmd = input("Введите команду (или exit): ")
    if cmd == "exit":
        break
    else:
        print(f"Выполняю команду {cmd}...")
```

## Задание 2. «Калькулятор 2.0»
Напишите программу, которая бесконечно спрашивает у пользователя два числа.

После ввода чисел программа должна вывести их сумму.

Сразу после этого цикл повторяется снова.

Единственный способ остановить программу — ввести слово "выход" вместо первого числа (нужна проверка условия и `break`).

```python
while True:
    cmd1  = input("Введите 1-ое число (или выход): ")
    if cmd1 == "выход":
        break
    else:
      cmd2  = input("Введите 2-ое число")
      result = float(cmd1) + float(cmd2)
      print("Ваш ответ", result)
```

## Часть 3. Вложенные циклы (Цикл внутри цикла)
Самая сложная тема для понимания. На каждом шаге внешнего цикла внутренний цикл прокручивается полностью.

**Аналогия:** Часы. Пока часовая стрелка (внешний цикл) делает 1 шаг, минутная (внутренний) делает 60 шагов.

**Пример (Таблица координат):**

```python
for x in range(1, 4):         # Внешний (строки)
    for y in range(1, 4):     # Внутренний (столбцы)
        print(f"Точка ({x}, {y})")
```

# Задание 3. «Таблица умножения»
Напишите программу, которая выводит таблицу умножения для чисел от 1 до 5. Формат вывода должен быть таким:` 1 * 1 = 1 1 * 2 = 2 ... 2 * 1 = 2 ... 5 * 5 = 25`

```python
for x in range(1, 6):
    for y in range(1, 6):
      result = float(x) * float(y)
      print(x, y , "=" , result)
```

## Часть 4. Итоговое задание
Задача: «Сортировка пузырьком» (`Bubble Sort`)

Это классическая алгоритмическая задача, которую спрашивают на собеседованиях. Она решается только через вложенные циклы.

Дан список: `numbers = [5, 2, 9, 1, 5, 6]`

**Алгоритм:**

1. Нужно много раз пройтись по списку слева направо.

2. Если левый элемент больше правого — их нужно поменять местами.

3. Самые большие элементы будут как «пузырьки» всплывать в конец списка.

**Задание:** Напишите код с двумя вложенными циклами, который отсортирует список numbers по возрастанию без использования `numbers.sort()`. Выведите отсортированный список.

**Подсказка:** чтобы поменять элементы местами в Python, используется синтаксис: `a, b = b, a`

```python
numbers = [5, 2, 9, 1, 5, 6]

print("Исходный список:", numbers)

n = len(numbers)

for i in range(n - 1):
    for j in range(n - 1 - i):
        if numbers[j] > numbers[j + 1]:
            numbers[j], numbers[j + 1] = numbers[j + 1], numbers[j]
print("Отсортированный список:", numbers)
```
